```python
k0 = 0.49 #thermal conductivity (cal/(s*cm*Ùè∞Ø`C))
delta_x = 2 #(cm)
delta_t = 0.1 #(s)
C = 0.2174 #heat capacity (cal/(g*`C))
dens = 2.7 #density (g/cm^3)
k = k0/C/dens #thermal diffusivity (cm^2/s)
lam = k*delta_t/(delta_x**2) 

#temperature of 6 position when t=0
T_time = [[100,0,0,0,0,50]]
```

```python
def Cal_T(T_point):
    """This function will calculate the temperature at the next timepoint, T_point need to be a list"""
    T_point_new = [100]
    for i in range(1,len(T_point)-1):
        new_T = T_point[i]+lam*(T_point[i+1]+T_point[i-1]-2*T_point[i])
        T_point_new.append(new_T)
    T_point_new.append(50)
    return T_time.append(T_point_new)
```

```python
#calculate the temperature at all timepoint till t=12(s)
for i in range(int(12/delta_t)):
    Cal_T(T_time[i])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
point = [0,2,4,6,8,10]
for i in [3,6,9,12]:
    plt.plot(point,T_time[int(i/delta_t)],'o-')
plt.xlim(0,10)
plt.legend(('t=3','t=6','t=9','t=12'))
plt.xlabel('Position on rod')
plt.ylabel('T')
```
